# **Análise da violência do Rio de Janeiro**

A violência no Brasil está apresentando níveis crescentes segundo o [**Altas da Violência**](http://www.ipea.gov.br/atlasviolencia/download/19/atlas-da-violencia-2019) do Instituro de Pesquisa Econômica Aplicada. Para se ter uma ideia, em 2017 o Brasil apresentou uma taxa de aproximadamente 31 mortes para cada 100 mil habitantes. 

No ano de 2018, o estado do Rio de Janeiro foi destaque nos noticiários devido ao aumento desenfreado nos níveis de violência. Foi então decretado a intervenção militar no estado. O § 2º do Art. 1º do Decreto 9.288 de 16 de fevereiro de 2018, diz que:

>"**O objetivo da intervenção é pôr termo a grave comprometimento da ordem pública no Estado do Rio de Janeiro.**".

>
O decreto transferiu o controle de todos os órgãos de segurança pública do estado para o Exército Brasileiro. A intervenção militar no estado do Rio de Janeiro teve seu encerramento no dia 31 de dezembro de 2018.

##Objetivo

#**Análises**
###Como visto anteriormente, o dataframe possui 56 variáveis. Por isso serão selecionadas apenas variáves relativas aos crimes de roubo, furto e homicídio.

A começar com roubo, furto e a recuperação de veículos.
Dessa forma, vamos selecionar apenas as colunas
* **furto_veiculos**
* **roubo_veiculo** 
* **recuperacao_veiculos**.

Para trabalhar isoladamentes estas três variáveis, será utilizado a função a função de agrupamento `groupby()` e será criado um novo **DataFrame** com o nome `df_2`. Os dados serão agrupados por ano ('vano') e mes ('mes') ordenados de forma crescente primeiramente por ano (1991 a 2019) em segundo lugar por meses (1 a 12).

```
# Agrupando os dados e criando um novo Data Frame
df_2 = (df.groupby(['vano', 'mes'])['furto_veiculos', 'roubo_veiculo', 'recuperacao_veiculos'].sum())
```

```
#Verificando o novo Data Frame
df_2.head()
```

## Porcentagem de veículos recuperados
A base de dados do Instituto de Segurança Pública do Estado do Rio de Janeiro informa apenas o número de carros que foram recuperados. Dessa forma, para encontrar a porcentagem de carros furtados e roubados que foram recuperados, será criada uma nova coluna denominada **porcent_veic_rec**.

```
#Cria inclui a coluna com a porcentagem de veículos furtados mais o veículos roubados que foram recuperados.
df_2['porcent_veic_rec'] = (((df_2['recuperacao_veiculos']) / (df_2['furto_veiculos'] + df_2['roubo_veiculo'])) *100)
```

```
#Mostra as primeiras 5 colunas do data frame
df_2.head()
```

###Análise de variação dos dados
Para verificar a amplitude de variação dos dados e a eventual presença de outliers (valores discrepantes). Será construído um gráfico do tipo **boxplot**.

```
#Constroe um gráfico do tipo boxplot
df_2.plot(kind='box', vert=False, figsize=(15,5), fontsize=14,patch_artist=True, cmap='cividis');
```

### Resultado
Analisando o boxplot é possível observar uma grande variação dos dados. Inclusive com alguns pontos sendo detectados como outliers. Em geral, os outtliers são excluídos das análises por causarem deturpações. Entretanto, como se trata de um registro de ocorrências e não se trata de uma análise experimental. Esses valores serão mantidos.


## Visualização dos dados de furtos, roubos e recuperação de veículos
Para termos uma visão da direcionamento (aumento ou diminuição) desses crimes. Serão construídos gráficos considerando número de ocorrências em com o passar dos anos.

```
#Cria subplots para as variáveis age e sex
fig, axs = plt.subplots(2, 2, sharex=False, figsize=(20,8))
 	 	 	

plt.subplot(2, 2, 1) #A utilização da função plt.subplot() precisa de coordenadas (número de linhas, número de colunas, índice)
plt.subplots_adjust(wspace=5) # Ajusta o espaço entre os gráficos
df_2.furto_veiculos.plot(grid=False, color='#EE0101') #define qual coluna (variável) será plotada no gráfico
plt.title('Furto de veículos', size=20) #Insere o título do gráfico
plt.ylabel('Número de ocorrências', size=14) #Insere o título do eixo y
plt.xlabel('Ano', size=14) #Insere o título do eixo x


plt.subplot(2, 2, 2)
plt.subplots_adjust(wspace=5)
df_2.recuperacao_veiculos.plot(grid=False, color='#062DCA')
plt.title('Recuperação de veículos', size=17)
plt.ylabel('Número de ocorrências', size=14)
plt.xlabel('Ano', size=14)


plt.subplot(2, 2, 3)
plt.subplots_adjust(wspace=5, hspace=2)
df_2.roubo_veiculo.plot(grid=False, color='#820707')
plt.title('Roubo de veículo', size=17)
plt.ylabel('Número de ocorrências', size=14)
plt.xlabel('Ano', size=14)


plt.subplot(2, 2, 4)
plt.subplots_adjust(wspace=5, hspace=2)
df_2.porcent_veic_rec.plot(grid=False, color='#2D06CA')
plt.title('Porcentagem de veículos recuperados', size=17)
plt.ylabel('%', size=14)
plt.xlabel('Ano', size=14)


plt.tight_layout()
```

#Resultados
O número de veículos furtados apresenta um tendência de declínio de 1991 a 2019. Por outro lado o número de veículos roubados aumentou consideravelmente. Após alcançar seus maiores valores em janeiro  e março de 2018, o número de veículos roubados vem apresentando um visível declínio.


O número de veículos recuperados também aumentou de maneira considerável. Note que de 1991 a 2000 a porcentagem de veículos recuperados era torno de 25%. Após o ano de 2000 este valor aumentou para uma média de 50% se mantendo até os dias atuais.



##Análise de furtos e roubos a transeuntes e de celulares.
Próximo passo será analisar os roubos a pessoas transitando além de roubo e furto de celulares. O roubo de celular começou a ser contabilizado em novembro de 2000, já os furtos inicou-se inclusão das ocorrências em 2003. 

Assim as análises para roubos será considerado janeiro de 2001 e furtos janeiro de 2003.

```
#Cria a lista apenas com as variáveis selecionadas para a criação da tabela estatística
df_3 = (df.groupby(['vano', 'mes'])['furto_transeunte', 'roubo_transeunte','roubo_celular', 'furto_celular'].sum())
```

```
#Cria subplots para as variáveis
fig, axs = plt.subplots(2, 2, sharex=False, figsize=(18,5))

plt.subplot(1, 2, 1)
plt.subplots_adjust(wspace=5, hspace=2)
df_rbc = (df_3.loc[2001:2019]['roubo_celular']) # Dados contabilizados a partir de 2001
df_rbc.plot(grid=False, color='#035F01') 
plt.title('Roubo celular de 2001 a 2019', size=17)
plt.ylabel('Número de ocorrências', size=14)
plt.xlabel('Ano', size=14)


plt.subplot(1, 2, 2)
plt.subplots_adjust(wspace=5, hspace=2)
df_ftc = (df_3.loc[2003:2019]['furto_celular']) # Dados contabilizados a partir de 2003
df_ftc.plot(grid=False, color='#03015F')
plt.title('Furto de celular de 2001 a 2019', size=17) 
plt.ylabel('Número de ocorrências', size=14)
plt.xlabel('Ano', size=14)


plt.tight_layout()
```

##Resultado
O crime de furto de transetunte apresentava números crescentes até o ano de 2016. Desde então, o número de ocorrências apresenta uma nítida redução. De maneira semelhante os roubos a transeunte também aumentaram, principalmente a partir de 07 de 2004. Há uma aparente tendência na redução no número de ocorrências. Principalmente se analisar o mês de fevereiro de 2018 onde foi registrado o recorde no número de ocorrências com 9669 registros. Contra os 6452 do mês de agosto de 2019.

As ocorrências de roubos de celulares apresentou uma redução de 1500 para 500 ocorrências nos primeiros anos de registro. Entretanto, a partir de 2013 o número de registros de roubo de celular aumentou veriginosamente. Da mesma forma, a ocorrências de furto de celular também aumentaram. 


```
fig, axs = plt.subplots(2, 2, sharex=False, figsize=(18,5))

plt.subplot(1, 2, 1) #A utilização da função plt.subplot() precisa de coordenadas (número de linhas, número de colunas, índice)
plt.subplots_adjust(wspace=5) # Ajusta o espaço entre os gráficos
df_3.furto_transeunte.plot(grid=False, color='#D0010B') #define qual coluna (variável) será plotada no gráfico
plt.title('Furto de transeunte', size=20) #Insere o título do gráfico
plt.ylabel('Número de ocorrências', size=14) #Insere o título do eixo y
plt.xlabel('Ano', size=14) #Insere o título do eixo x

plt.subplot(1, 2, 2)
plt.subplots_adjust(wspace=5)
df_3.roubo_transeunte.plot(grid=False, color='#065197')
plt.title('Roubo de transeunte', size=17)
plt.ylabel('Número de ocorrências', size=14)
plt.xlabel('Ano', size=14)

plt.tight_layout()
```

##Resultados
Tanto número de ocorrências de furto e roubo de transeunte aumentaram desde o início dos registros. Entretanto, o gráfico de furto de transeunte apresenta vários picos consecutivos. 

A tabela baixo traz os dados relativos ao furto de transeunte. Os registros foram selecionados a partir do ano de 2005 (início dos picos) com os meses de janeiro, fevereiro e março. 

Note que esses picos acontecem no mês de fevereiro ou no mês de março, coincidindo com o carnaval. Sabe-se que o Rio de Janeiro é um destino bastante procurado para neste período. O que ajuda a entender esta variação nos dados.
